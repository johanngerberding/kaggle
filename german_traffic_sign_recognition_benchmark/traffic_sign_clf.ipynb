# Data Preprocessing

```python
def preprocess_images(df, base_dir, resize=(30, 30), brg=1, normalize=True):
    data = []
    labels = []
    for i in df.index:
        img_path = os.path.join(base_dir, df.loc[i, 'Path'])
        img = cv2.imread(img_path, brg)
        img = cv2.resize(img, (30, 30))
        labels.append(df.loc[i, 'ClassId'])
        data.append(img)
    
    y = np.stack(labels, axis=0)
    X = np.stack(data, axis=0)
    
    if normalize:
        X = X/255.
    
    return X, y
```

```python
X_train, y_train = preprocess_images(train_data, base_dir)
X_train.shape
```

```python
y_train.shape
```

```python
plt.figure(figsize=(14,6))
plt.hist(y_train, 42) # 42 different classes
plt.title('Histogram of training data labels')
plt.xticks(range(42))
plt.show()
```

```python
os.chdir(base_dir)
test_data = pd.read_csv('Test.csv')
test_data.head()
```

```python
X_test, y_test = preprocess_images(test_data, base_dir)
X_test.shape
```

```python
y_test.shape
```

```python
plt.figure(figsize=(14,8))
plt.hist(y_test, 42)
plt.title('Histogram of test traffic sign images')
plt.xticks(range(42))
plt.show()
```

As you can see, we are dealing with a **scewed dataset**. Some classes are much more frequent then others. This has an impact on the **performance measure** for our classifier.

# Building a model

```python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(30, 30, 3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Dropout(0.25))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(43, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=20,
                   validation_data=(X_test, y_test))
```
